```python
# import numpy as np
# import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# import matplotlib.pyplot as plt


data_df = pd.read_csv("./AmesHousing.csv")

x = data_df[['Lot Area', 'Overall Qual', 'Year Built', 'Gr Liv Area']]
y = data_df['SalePrice']   

x = x.dropna()
y = y[x.index]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42 )

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

sorted_idx = x_test['Lot Area'].argsort()

plt.figure(figsize=(8,6))
plt.scatter(x_test['Lot Area'], y_test, color="blue", label="Actual SalePrice")
plt.plot(x_test['Lot Area'].iloc[sorted_idx], y_pred[sorted_idx], color="red", linewidth=2, label="Predicted")
plt.xlabel("Lot Area")
plt.ylabel("SalePrice")
plt.title("Linear Regression Model")
plt.legend()
plt.show()

print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.show()
```

```python
errors = y_test - y_pred

sns.histplot(errors, kde=True)
plt.title("Error Distribution")
plt.show()
```

```python
# import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
# import pandas as pd 

# Create a figure and axes
fig, axs = plt.subplots(3, 2, figsize=(14, 12))
fig.suptitle('Histograms with KDE of Various Distributions', fontsize=16)

data_df = pd.read_csv("./AmesHousing.csv")
real_data = data_df['SalePrice'].dropna()
# real_data = data_df['Year Built'].dropna()
# real_data = data_df['Lot Area'].dropna()

# 1. Normal Distribution

sns.histplot(real_data, kde=True, ax=axs[0, 0], color='blue', stat='density')
axs[0, 0].set_title('Normal Distribution')
axs[0, 0].grid(True)

# 2.Skewed Distribution

#Right skewed histogram
sns.histplot(real_data,  kde=True, ax=axs[0,1], color='red', stat='density', bins=30, label = 'Right Skewed')
# Left Skewed histogram
sns.histplot(-real_data, kde=True, ax=axs[0,1], color='green', stat='density', bins=30, label = 'Left Skewed')
axs[0,1].set_title('Skewed Distribution')
axs[0,1].legend()
axs[0,1].grid(True)

# 3. Uniform Distributionlow
data_uniform = (real_data - real_data.min())/ (real_data.max() - real_data.min())# uniform distribution betweeen 0 and 1
sns.histplot(data_uniform, kde=True, ax=axs[1,0], color='purple', stat='density', bins=30)
axs[1,0].set_title('Uniform Distribution')
axs[1,0].grid(True)

#4. Bimodel Distribution
half = len(real_data)//2
data_bimodel = np.concatenate([real_data[:half],
                               real_data[half:]])
sns.histplot(data_bimodel, kde=True, ax=axs[1,1], color='orange',stat='density', bins=30)
axs[1,1].set_title('Bimodel Distribution')
axs[1,1].grid(True)
    
#5. multimodel distribution

data_multimodel = np.concatenate([real_data[::3],
                                  real_data[1::3],
                                  real_data[2::3]
])
sns.histplot(data_multimodel, kde=True, ax=axs[2,0], color='pink',stat='density', bins=30)
axs[2,0].set_title('multimodel Distribution')
axs[2,0].grid(True)

fig.delaxes(axs[2,1])

plt.tight_layout()
plt.show()
```
